```python
# all images are load in variable x...
Arq =['Cravo.zip','Gerânio.zip','Margaridas.zip','Orquídeas.zip',
      'Rosas.zip']

img_name=[]
x_gray=[]
x_original = []
for FILE in Arq:
  file_name = zipfile.ZipFile(FILE, 'r')
  file_name.extractall()
  with zipfile.ZipFile(FILE, "r") as f:
      for name in f.namelist():
          img_name.append(name)
```

```python
y_names = []
for yy in img_name:
  name = yy[0]+yy[1]+yy[2]
  y_names.append(name)
```

```python
unique_list = []    
for yy in y_names:
  if yy not in unique_list:
    unique_list.append(yy)
```

```python
n = len(unique_list)
y = []
for yy in y_names:
  for i in range(n):
    if(yy == unique_list[i]):
      y.append(i)
```
